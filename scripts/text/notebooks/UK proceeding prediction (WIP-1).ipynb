```python
import pandas as pd
import numpy as np
from sklearn.externals import joblib
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
import nltk
from nltk import word_tokenize          
from nltk.stem.porter import PorterStemmer
import re
import string

stemmer = PorterStemmer()
def stem_tokens(tokens, stemmer):
    stemmed = []
    for item in tokens:
        stemmed.append(stemmer.stem(item))
    return stemmed

#def tokenize(text):
#    tokens = nltk.word_tokenize(text)
#    stems = stem_tokens(tokens, stemmer)
#    return stems

def tokenize(text):
    text = "".join([ch for ch in text if ch not in string.punctuation])
    tokens = nltk.word_tokenize(text)
    stems = stem_tokens(tokens, stemmer)
    return stems
```

```python
vect_name = '../models/vec_count_23gram_083.joblib'
clf_name = '../models/clf_elasticnet_083.joblib'
```

```python
vect = joblib.load(vect_name)
clf = joblib.load(clf_name)
```

```python
len(vect.vocabulary_)
```

```python
df = pd.read_csv('../uk_proceedings_clean.zip', usecols=['PARTY', 'SPEECH_TEXT'])
```

```python
sdf = df[df.PARTY.isin(['Labour', 'Conservative']) & df.SPEECH_TEXT.notnull() & (df.SPEECH_TEXT.str.len() >= 2000)]
sdf
```

```python
X = vect.transform(sdf.SPEECH_TEXT)
tfidf = TfidfTransformer()
X = tfidf.fit_transform(X)
y_pred = clf.predict(X)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(sdf.PARTY, y_pred)
```

```python
sdf['PREDICT'] = y_pred
sdf
```

```python
sdf.columns = ['true', 'text', 'predicted']
#sdf.to_csv('../data/uk_proceedings_prediction.csv', index=False, columns=['true', 'predicted'])
```
