
1. Summarize Media Data:

   Go to uk_media_clean.csv.zip and create a table that has following columns:
   src_name, n_transcripts, from_date, to_date


```python
import pandas as pd
import numpy as np
```

```python
#df = pd.read_csv('../media_text/uk_media_clean_10k.csv', usecols=['pubDate', 'language', 'title', 'sourceName', 'url', 'articleId'], parse_dates=['pubDate'], infer_datetime_format=True)
df = pd.read_csv('../media_text/uk_media_clean_deduped.csv', parse_dates=['pubDate'], infer_datetime_format=True)
df
```

```python
df = df[df.pubDate >= '2010-01-01']
```

```python
sdf = df.groupby('sourceName').agg({'pubDate': [np.min, np.max], 'sourceName': 'count'})
sdf.reset_index(inplace=True)
sdf.columns = ['src_name', 'n_transcripts', 'from_date', 'to_date']
sdf
```

```python
#sdf.to_csv('../data/uk_media_clean_summarized.csv', index=False)
```

## Strip white space and lower case

```python
df['src_name'] = df.sourceName.str.strip().str.lower()
```

## Take out duplicate flags

```python
df.loc[:, 'src_name'] = df['src_name'].str.replace('\s*\(dup\??\)', '')
```

## Ad-hoc search and replace

```python
ad_hocs = """
'','
&,and
al-sharq al-awsat,asharq al-awsat
asharq alawsat,asharq al-awsat
asharq al-awsatt,asharq al-awsat
northampton and chronicle echo,northampton chronicle and echo
hartlepool mail,hartle pool mail
"""
```

```python
cdf = df
for a in ad_hocs.split('\n'):
    if a.strip() != '':
        s, r = a.split(',')
        print("Search '{0:s}' and replace with '{1:s}'".format(s, r))
        cdf.loc[:, 'src_name'] = cdf['src_name'].str.replace(s, r)
```

## Grouping and summarized

```python
sdf = cdf.groupby('src_name').agg({'pubDate': [np.min, np.max], 'src_name': 'count'})
sdf.reset_index(inplace=True)
sdf.columns = ['src_name', 'n_transcripts', 'from_date', 'to_date']
sdf
```

```python
sdf.to_csv('../data/uk_media_clean_summarized_deduped_201x.csv', index=False)
```
