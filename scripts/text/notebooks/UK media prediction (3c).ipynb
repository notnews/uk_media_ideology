```python
import pandas as pd
import numpy as np
from sklearn.externals import joblib
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
vect_name = '../models/vec_count_ukmedia_23gram_074.joblib'
clf_name = '../models/clf_elasticnet_ukmedia_23gram_074.joblib'
```

```python
vect = joblib.load(vect_name)
clf = joblib.load(clf_name)
```

```python
len(vect.vocabulary_)
```

```python
df = pd.read_csv('../media_text/uk_media_clean_deduped_201x_gt1k.csv', usecols=['pubDate', 'src_name', 'text'])
df
```

```python
X = vect.transform(df.text.astype(str))
```

```python
X.shape
```

```python
tfidf = TfidfTransformer()
X = tfidf.fit_transform(X)
```

```python
X.shape
```

```python
y_pred = clf.predict(X)
```

```python
y_pred.shape
```

```python
X_proba = clf.predict_proba(X)
```

```python
X_proba
```

```python
proba_cons = X_proba[:,0]
```

```python
df['predicted'] = y_pred
df['pred_cons'] = proba_cons
df
```

```python
df.columns
```

```python
df.columns = ['date', 'text', 'news_source', 'predicted', 'pred_cons']
#df.to_csv('../data/uk_media_prediction_3c.csv', index=False, columns=['date', 'news_source', 'pred_cons'])
```

sdf = df.groupby('news_source').agg({'pred_cons': np.mean})
